### Import HappyGeneration

```python
from happytransformer import HappyGeneration
```

### Create a HappyGeneration Object 

```python
happy_gen = HappyGeneration("GPT-NEO", "EleutherAI/gpt-neo-1.3B")
```

## happy_gen.generate_text()

```python
result = happy_gen.generate_text("We must invest in ")
```

```python
print(result)
```

```python
print(result.text)
```

### import GENSettings

```python
from happytransformer import GENSettings
```

### Greedy

```python
greedy_settings = GENSettings(no_repeat_ngram_size=2,  max_length=50)
```

```python
result_greedy = happy_gen.generate_text("We must invest in ", args=greedy_settings)
```

```python
print(result_greedy)
```

```python
print(result_greedy.text)
```

### Generic Samping

```python
generic_sampling_settings = GENSettings(do_sample=True, top_k=0, temperature=0.5,  max_length=50)
```

```python
result_generic_sampling = happy_gen.generate_text("We must invest in ", args=generic_sampling_settings)
```

```python
print(result_generic_sampling.text)
```

### Top-k Sampling

```python
top_k_sampling_settings = GENSettings(do_sample=True, top_k=15, temperature=0.5,  max_length=50, no_repeat_ngram_size=2)
```

```python
result_top_k_sampling = happy_gen.generate_text("We must invest in ", args=top_k_sampling_settings)
```

```python
print(result_top_k_sampling.text)
```
